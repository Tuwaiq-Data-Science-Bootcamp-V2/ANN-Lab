```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder,StandardScaler,MinMaxScaler
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
data=pd.read_csv("Bank_Customers.csv")
```

```python
data
```

```python
data = data.drop(columns=['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data
```

```python
data.describe()
```

```python
transformers={}
mm=['CreditScore','Age','Tenure','NumOfProducts','EstimatedSalary']

for col in mm:
    scaler=MinMaxScaler()
    scaler.fit(data[col].values.reshape(-1, 1))
    transformers[col] = scaler
    data[col] = scaler.transform(data[col].values.reshape(-1, 1))
```

```python
data
```

```python
scaler=StandardScaler()
scaler.fit(data['Balance'].values.reshape(-1, 1))
transformers['Balance']=scaler
data['Balance']=scaler.transform(data['Balance'].values.reshape(-1, 1))
```

```python
data
```

```python
mm=['Geography','Gender','HasCrCard','IsActiveMember']
for col in mm:
    encoder=OneHotEncoder()
    encoder.fit(data[col].values.reshape(-1, 1))
    transformers[col] = encoder
    arr = encoder.transform(data[col].values.reshape(-1, 1)).toarray()
    for i in range(arr.shape[1]):
        data[col+'_'+str(i)] = arr[:,i]
```

```python
data
```

```python
data=data.drop(columns=mm,axis=1)
```

```python
data
```

```python
data['Exited'].value_counts()
```

```python
X=data.drop('Exited',axis=1).to_numpy()
y=data['Exited'].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=42)
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(256,activation='relu',input_shape=(15,)))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(64,activation='relu'))
model.add(tf.keras.layers.Dense(32,activation='relu'))
model.add(tf.keras.layers.Dense(16,activation='relu'))
model.add(tf.keras.layers.Dense(8,activation='relu'))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))         
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam',
              loss = tf.keras.losses.BinaryFocalCrossentropy(apply_class_balancing=True,alpha=0.8),
              metrics = ['accuracy'])
              
```

```python
model.fit(X_train,y_train,validation_data=(X_val,y_val),batch_size=32,epochs=150)
```
